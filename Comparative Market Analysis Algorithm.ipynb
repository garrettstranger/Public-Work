# Strategy

In order to measure how succesful this model is in estimating house prices, mean absolute error will be used. With all models, the goal should be minimize or maximize a particular measurement variable and in this case it will be the mean absolute error. This is defined as the average of the absolute errors. The reason for this selection is because its hard to gain intuition around the mean square, whereas absolute mean error allows us to say the model was off by $50,000 or similar on average.

In terms of the approach for unit testing and functional testing, this dataset was stratified into random chunks to sample the absolute mean error. ** The proprietary code used to do this has been removed **
